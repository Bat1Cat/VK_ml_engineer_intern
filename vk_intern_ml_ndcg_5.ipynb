```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
qid = X_train['query_id'].values
```

```python
X_train_reset_index = X_train.reset_index(drop=True)
y_train_reset_index = y_train.reset_index(drop=True)
qid_reset_index = np.arange(len(qid))

sorted_indices = np.argsort(qid)
X_train_sorted = X_train.iloc[sorted_indices]
y_train_sorted = y_train.iloc[sorted_indices]
qid_sorted = qid[sorted_indices]
```

```python
qid_test = X_test['query_id'].values
```

```python
X_test_reset_index = X_test.reset_index(drop=True)
y_test_reset_index = y_test.reset_index(drop=True)
qid_reset_index = np.arange(len(qid_test))

sorted_indices = np.argsort(qid_test)
X_test_sorted = X_test.iloc[sorted_indices]
y_test_sorted = y_test.iloc[sorted_indices]
qid_test_sorted = qid_test[sorted_indices]
```

```python
numerical_cols = [cname for cname in X_train_sorted.columns if X_train_sorted[cname].dtype in ['int64', 'float64']]
```

```python
numerical_cols
```

```python
numerical_transformer = StandardScaler()
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
    ])
```

```python
import xgboost as xgb
from sklearn.metrics import ndcg_score

full_pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('xgboost', xgb.XGBRanker(tree_method="hist", objective="rank:ndcg"))
])
```

```python
full_pipeline.fit(X_train_sorted, y_train_sorted, xgboost__qid=qid_sorted)


y_pred = full_pipeline.predict(X_test_sorted)
```

```python

ndcg_5 = ndcg_score(np.expand_dims(y_test_sorted.values, axis=0), np.expand_dims(y_pred, axis=0), k=5)

ndcg_3 = ndcg_score(np.expand_dims(y_test_sorted.values, axis=0), np.expand_dims(y_pred, axis=0), k=3)


print("NDCG@5 на тестовом наборе данных:", ndcg_5)
print("NDCG@5 на тестовом наборе данных:", ndcg_3)
```
